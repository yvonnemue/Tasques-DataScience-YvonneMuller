```python
#Exercici 1 
#Crea un np.array d'una dimensió, que inclogui l'almenys 8 nombres sencers, data type int64. 
#Mostra la dimensió i la forma de la matriu.

import numpy as np

prim_arr = np.array([1,4,6,8,3,44,5,66,7])
print(type(prim_arr[5]))
print('Dimensió i forma:', prim_arr.shape) #Una sola dimensió de 9 elements

seg_arr = np.array([[1,4,6,8,3,44,5,66,7],[1,4,6,8,3,44,5,66,7]])
print(type(seg_arr[0][5]))
print('Dimensió i forma:', seg_arr.shape) #Dues dimensions de 9 elements cadascuna
```

```python
#Exercici 2 

#De la matriu de l'exercici 1, calcula el valor mitjà dels valors introduïts i 
#resta la mitjana resultant de cada un dels valors de la matriu.

print('Mitjana', np.mean(prim_arr))
print('Valors de l\'array menys la mitjana', prim_arr-np.mean(prim_arr))
```

```python
#Exercici 3
#Crea una matriu bidimensional amb una forma de 5 x 5. 
#Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

ter_arr = np.array([[1,4,6,3,4],[6,8,44,3,22],[3,44,6,33,21],[5,6,44,66,2],[75,33,6,3,0]])
print('Forma:', np.shape(ter_arr))
print('Máx. num. array:', np.max(ter_arr))
print('Máx. num. axis:', np.max(ter_arr,axis=1))
```

```python
#Exercici 4 

#Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : 
#"les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o 
#si una de les matrius té una mida d'1". 

broadcast_1 = ter_arr + 2 #Suma 2 a tots els valors de l'array
print('Arr original: \n', ter_arr)
print('Arr + 2: \n', broadcast_1) 

A = np.array([21, 22, 23])
B = np.array([11, 12, 13])
broadcast_2 = A - B #El resultat es un array de tres elements: la resta del primer element de cada array, la resta del segon y la resta del tercer.
print('Resta de dos arrays amb la mateixa mida: \n', broadcast_2)

C = np.array([[21, 22],[0,2]])
D = np.array([[10, 20],[4,0]])
broadcast_3 = C + D
print('Suma de dos arrays amb la mateixa mida (2x2): \n',broadcast_3)
```

```python
#Exercici 5 

#Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. 
#I suma els seus valors.

Valor_1 = ter_arr[3][3]
Valor_2 = ter_arr[1][4]
Resta = Valor_1 - Valor_2
print('Valor 1: ', Valor_1, '\nValor 2: ', Valor_2,'\nResta: ', Resta)

Valor_col = ter_arr[:,3]
Valor_fila =ter_arr[3]
Suma = Valor_col + Valor_fila 
print('Valor columna: ', Valor_col, '\nValor fila: ', Valor_fila,'\nSuma: ', Suma)
```

```python
#Exercici 6 

#Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat, agafant cada element 
#i comprovant si es divideix uniformement per quatre.

#Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

mask = (ter_arr %4 == 0)
print(mask)
```

```python
#Exercici 7 

#A continuació, utilitzeu aquesta màscara per indexar a la matriu de números original. 
#Això fa que la matriu perdi la seva forma original, reduint-la a una dimensió, 
#però encara obteniu les dades que esteu cercant.

print('Valors divisibles per 4: \n', ter_arr[mask]) #Els 7 valors divisibles per 4 de la matriu anterior
```
