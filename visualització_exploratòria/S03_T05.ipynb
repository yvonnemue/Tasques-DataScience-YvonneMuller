```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

pd.options.display.max_columns = None
pd.options.display.max_rows = 100

data = pd.read_csv('DelayedFlights.csv')

#Crear el mateix sample del dataset per a que le simatges tinguin sentit
data = data.sample(n = 10000, random_state=30)
```

```python
data.info()
```

```python
data_rel = data[['FlightNum','UniqueCarrier','Month','DepTime','CRSDepTime','ArrTime','CRSArrTime','ActualElapsedTime','CRSElapsedTime','AirTime','Origin','Dest','Distance']]
```

```python
data_rel.info()
```

```python
#Visualització exploratòria
img = mpimg.imread('fig1.png')
img2 = mpimg.imread('fig2.png')

imgplot = plt.imshow(img)
plt.axis('off')
plt.title('Distribució d\'aerolínies')
plt.show()

imgplot2 = plt.imshow(img2)
plt.axis('off')
plt.title('Distribució de l\'endarreriment')
plt.show()
```

### Exercici 2

```python
data_rel.head(10)
```

```python
#Núm de missing values per columna
data_rel.isnull().sum()
```

```python
#Substituir missing values per 0 i convertir columnes a int
```

```python
data_rel['ArrTime'] = data_rel['ArrTime'].fillna(0.0).astype(int)
data_rel['ActualElapsedTime'] = data_rel['ActualElapsedTime'].fillna(0.0).astype(int)
data_rel['CRSElapsedTime'] = data_rel['CRSElapsedTime'].fillna(0.0).astype(int)
data_rel['AirTime'] = data_rel['AirTime'].fillna(0.0).astype(int)
data_rel['DepTime'] = data_rel['DepTime'].astype(int)
```

```python
data_rel.info()
```

```python
data_rel.head(10)
```

```python
#Resum estadístic de les columnes
data_rel.describe().apply(lambda s: s.apply('{0:.3f}'.format))
```

```python
#Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

#Velocitat mitjana de vol (milles per hora)
data_rel['Vel_mitjana_vol(mi/h)'] = ((data_rel['Distance']/data_rel['ActualElapsedTime'])*60).round(2)

#Minuts de retard
data_rel['Retard(min)'] = data_rel['ArrTime'] - data_rel['CRSArrTime']
data_rel['Retard(min)'] = data_rel['Retard(min)'].apply(lambda x: "0" if x < 0 else x).astype(int)
```

```python
data_rel.head(10)
```

```python
data_rel.info()
```

```python
#Taula de les aerolínies amb més endarreriments acumulats
aero_retard = data_rel[['UniqueCarrier','Retard(min)']].groupby('UniqueCarrier').sum().sort_values(by=('Retard(min)'), ascending=False)
print(aero_retard)

img3 = mpimg.imread('fig3.png')
imgplot3 = plt.imshow(img3)
plt.axis('off')
plt.title('Mitjana d\'endarreriment per aerolínia')
plt.show()

img4 = mpimg.imread('fig4.png')
img5 = mpimg.imread('fig5.png')
img6 = mpimg.imread('fig6.png')


imgplot4 = plt.imshow(img4)
plt.axis('off')
plt.title('Endarreriment en la sortida y l\'arribada')
plt.show()

imgplot5 = plt.imshow(img5)
plt.axis('off')
plt.title('Endarreriment en la sortida y l\'arribada per aerolínia')
plt.show()

imgplot6 = plt.imshow(img6)
plt.axis('off')
plt.title('Endarreriment en la sortida y l\'arribada per aerolínia')
plt.show()
```

```python
#Quins són els vols més llargs? 
#Per milles
vol_llarg_milles = data_rel[['FlightNum','Origin','Dest','Distance']].groupby(['FlightNum','Origin','Dest']).max().sort_values(by=('Distance'), ascending=False).head(50)
vol_llarg_milles
```

```python
#Per temps de vol
vol_llarg_temps = data_rel[['FlightNum','Origin','Dest','AirTime']].groupby(['FlightNum','Origin','Dest']).max().sort_values(by=('AirTime'), ascending=False).head(50)
print(vol_llarg_temps)

img7 = mpimg.imread('fig7.png')

imgplot7 = plt.imshow(img7)
plt.axis('off')
plt.title('Endarreriments i temps de vol per aerolínia')
plt.show()
```

```python
#I els més endarrerits?
vols_endarrerits = data_rel[['FlightNum','Origin','Dest','Retard(min)']].groupby(['FlightNum','Origin','Dest']).max().sort_values(by=('Retard(min)'), ascending=False).head(50)
vols_endarrerits 
```
